## Data Engineer Interview - Quiz 3

### Challenge: Healthcare – Patient Visit Modeling
### A health-tech startup wants to track patient records, visits, prescriptions, and lab results, with strict privacy history. **Question: **Design an ER diagram and briefly explain how you’d enable audit/history tracking and HIPAA compliance in your model.



Healthcare – Patient Visit Modeling Solution by Dimensional Model

1. fact_patient_visits

- visit_key (PK)
- patient_key (FK)
- provider_key (FK)
- visit_date_key
- visit_type
- visit_count (always 1)

2. fact_prescriptions

- prescription_key (PK)
- visit_key (FK)
- medication_key (FK)
- dosage
- prescribed_qty

3. fact_lab_results

- lab_result_key (PK)
- visit_key (FK)
- lab_test_key (FK)
- result_value

4. dim_patient

- patient_key (PK)
- gender
- dob
- effective_from
- effective_to
- is_current

5. dim_provider

- provider_key (PK)
- specialty

6. dim_medication

- medication_key (PK)
- medication_name

7. dim_lab_test

- lab_test_key (PK)
- test_name
- unit

8. dim_date

- date_key (PK)
- date
- month
- year

Audit/History Tracking

- Use Slowly Changing Dimensions (Type 2) for dim_patient to preserve demographic history.
- Fact tables -> append-only — no updates or deletes.
- Corrections to prescriptions or lab results create new fact rows with updated values.

HIPAA Compliance

- Store PII (name, address, SSN) in a separate secured dimension (dim_patient_identity) with encryption.
- Enable row-level security and log access in an audit table. ✅
