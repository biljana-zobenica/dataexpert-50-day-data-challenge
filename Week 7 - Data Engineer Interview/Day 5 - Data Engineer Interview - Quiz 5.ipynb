## Data Engineer Interview - Quiz 5

### Education Platform – Course Completion
### A bootcamp wants to analyze completion rates for courses across different student cohorts.
### **Question: Design the core data model to capture enrollments, assignments, submissions, grading, and completion tracking.

Education Platform – Course Completion Solution by Dimensional Modeling

1. fact_enrollments

- enrollment_key (PK)
- student_key (FK → dim_student)
- course_key (FK → dim_course)
- cohort_key (FK → dim_cohort)
- enroll_date_key (FK → dim_date)
- completion_status
- completion_date_key (FK → dim_date)

2. fact_assignment_submissions

- submission_key (PK)
- student_key (FK → dim_student)
- assignment_key (FK → dim_assignment)
- course_key (FK → dim_course)
- submission_date_key (FK → dim_date)
- score
- is_passed (bolean)

3. dim_student

- student_key (PK)
- cohort_key (FK → dim_cohort)
- enrollment_type

4. dim_course

- course_key (PK)
- course_name
- duration_weeks

5. dim_assignment

- assignment_key (PK)
- course_key (FK → dim_course)
- assignment_type
- max_score

6. dim_cohort

- cohort_key (PK)
- start_date
- end_date

7. dim_date

- date_key (PK)

Completion Tracking

- Course completion derived from fact_enrollments.completion_status.
- Assignment completion and grading tracked at submission level.
- Completion rates calculated by aggregating enrollments by cohort and course. ✅
