## Data Modeling - Data Modeling

### 1 - Your company’s loyalty program database tracks customers who frequently update their contact details such as phone, email, and address over multiple years. To support marketing and compliance audits, you need a data model that preserves the complete history of these changes while enabling fast access to current information. How would you design the data model and schema? How would you query the latest versus historical information? Also, describe your dimensional model schema design and discuss the trade-offs involved.

For explained use case, I would model customer contact details data by using a Slowly Changing Dimension Type 2 (SCD-2) table. 
In this table every change would create a new versioned record with effective dates and a current-flag. When we are querying the latest data, we would use the condition when: is_current = true, while for historical audits we would use date ranges from history.
A clear, existing trade-off: In the dimensional model, we have accurate and virsioned historical reporting, but with increased table size and potentially more complex ETL logic. ✅
