## Data Modeling - Core Data Modeling Elements

### 1 - What distinguishes a fact from a dimension in data modeling?

Dimensions are events that happen over time, while facts are static attributes.

Facts are aggregated and stored in logs, while dimensions are attributes from a snapshot of state. ✅

Dimensions come from logs, while facts come from a snapshot of state.

Facts are non-numeric and dimensions are numeric.

### 2 - What does bucketization help achieve when modeling data dimensions?

It allows for clearer data analysis by categorizing similar values into groups. ✅

It increases the cardinality of dimensions.

It is used to label dimensions as facts.

It reduces the number of unique keys in fact tables.

### 3 - How can a dimension like 'Dim is active' be considered both a fact and a dimension?

It can be derived from aggregated facts even though it appears as an attribute. ✅

A dimension is always a subset of a fact.

It depends on the user's perspective.

Dimensions and facts are interchangeable in data models.

### 4 - Why is the dimension definition important in long-term data modeling?

They hold no real significance in data reporting.

Dimensions are defined based on temporary events.

Once a dimension is defined, it is easy to change.

Re-defining dimensions can be costly and time-consuming. ✅

### 5 - What role does cardinality play in distinguishing between facts and dimensions?

Cardinality is irrelevant to fact and dimension distinction.

Facts have high cardinality because they represent many actions/events over time. ✅

High cardinality dimensions are preferred in all data models.

Dimensions always have lower cardinality than facts.
