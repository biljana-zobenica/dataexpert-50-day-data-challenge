## Data Engineer Interview - Quiz 4

### Case: Product Catalog with Multi-Categories
### An online marketplace allows products to belong to multiple categories, and categories can have subcategories. **Question: **What schema would you use to support flexible product categorization, fast lookups, and hierarchy traversal?

Product Catalog with Multi-Categories Modeling Solution by Dimensional Model

1. fact_product_category

- product_key (FK)
- category_key (FK)
- assignment_count (always 1)

One row per product–category assignment (many-to-many).

2. dim_product

- product_key (PK)
- product_name
- brand

3. dim_category

- category_key (PK)
- category_name
- parent_category_key (PK)
- effective_from
- effective_to
- is_current (boolean)

Hierarchy Support

Category hierarchy stored in dim_category via parent_category_key.
Optionally add a category bridge (closure) table:
- ancestor_category_key
- descendant_category_key
- depth
for fast rollups across subcategories.

Fast Lookups

Star schema enables fast joins from fact to dimensions.
Precompute category paths for analytics queries. ✅
