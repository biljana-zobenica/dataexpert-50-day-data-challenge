### 3 - Explain the importance of "giving the model time to think" in prompt engineering.

The prompts need to be structured properly, including the key elements of an effective prompt, such as: context, role, task (or instruction) and format. Giving the model time to think allows it to go through a problem step by step instead of rushing to providing an answer. This is important as the model processes each element sequentially and applies better logical reasoning, improving accuracy and quality of its final response. ✅

### 4 - What is a recommended best practice when crafting prompts?

Breaking complex tasks into simpler subtasks ✅

Ignoring the model’s previous outputs

Using complex language to challenge the model

Writing vague and open-ended instructions
